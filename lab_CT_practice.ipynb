```python
import numpy as np 
import matplotlib.pyplot as plt 



# Task 1: Signal Smoothing with First-Order Difference Equation
def difference_equation(x, alpha):
    """
    Apply the first-order difference equation to smooth the signal.
    
    Parameters:
    x (array-like): Input signal.
    alpha (float): Smoothing factor (0 < alpha < 1).
    
    Returns:
    array: Smoothed signal.
    """
    y = np.zeros_like(x)
    y[0] = x[0]  # Initial condition
    for n in range(1, len(x)):
        y[n] = alpha * x[n] + (1 - alpha) * y[n - 1]
    return y

# Input signal
x = np.array([1, 2, 2, 10, 2, 2, 1])  # Example input signal

alpha = np.array([0.2, 0.8])  # Smoothing factors

# Apply the difference equation for each alpha
#smoothed_signals = [difference_equation(x, a) for a in alpha]
# Plotting the results

smooth_signal1 = difference_equation(x, alpha[0])
smooth_signal2 = difference_equation(x, alpha[1])


N = 100;
n = np.arange(N);
clean_signal = np.sin(2*np.pi*n/N)
noise_signal = clean_signal + np.random.randn(N)


smooth_signal3= difference_equation(noise_signal, alpha[0])
smooth_signal4 = difference_equation(noise_signal, alpha[1])



plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(smooth_signal1, label=f'Alpha = {alpha[0]}')
plt.plot(smooth_signal2,label=f'Alpha = {alpha[1]}')
plt.plot(x, 'k--', label='Original Signal', alpha=0.5)
plt.title('Signal Smoothing using First-Order Difference Equation')

plt.xlabel('n')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(smooth_signal3, label=f'Alpha = {alpha[0]}')
plt.plot(smooth_signal4, label=f'Alpha = {alpha[1]}')
plt.plot(noise_signal, 'k--', label='Original Signal', alpha=0.8)
plt.title('Signal Smoothing with Noise using First-Order Difference Equation')

plt.xlabel('n')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()



plt.figure(figsize=(10, 6))
plt.plot(noise_signal, label='Original Signal with Noise', alpha=0.5)
for a in alpha:
    plt.plot(difference_equation(noise_signal,a),label=f'Alpha ={a}')
```
