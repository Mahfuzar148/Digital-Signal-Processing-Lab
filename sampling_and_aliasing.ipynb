````python
import numpy as np
import matplotlib.pyplot as plt

# Original continuous-time signal
f_signal = 10  # signal frequency in Hz
t_cont = np.arange(0, 1, 0.00001)  # fine time grid for continuous signal
x_cont = np.sin(2 * np.pi * f_signal * t_cont)

# Sampling rates
fs_high = 50  # High sampling rate (above Nyquist)
fs_low = 15   # Low sampling rate (below Nyquist for this signal)

# Sampled signals
t_high = np.arange(0, 1, 1/fs_high)
x_high = np.sin(2 * np.pi * f_signal * t_high)

t_low = np.arange(0, 1, 1/fs_low)
x_low = np.sin(2 * np.pi * f_signal * t_low)

# Plotting
plt.figure(figsize=(12, 6))

# Original and high-rate sampling
plt.subplot(2, 1, 1)
plt.plot(t_cont, x_cont, label='Original Signal')
plt.stem(t_high, x_high, linefmt='r-', markerfmt='ro', basefmt=' ', label='High-rate Samples')
plt.title('Sampling above Nyquist Rate')
plt.legend()

# Original and low-rate sampling (Aliasing)
plt.subplot(2, 1, 2)
plt.plot(t_cont, x_cont, label='Original Signal')
plt.stem(t_low, x_low, linefmt='g-', markerfmt='go', basefmt=' ', label='Low-rate Samples (Aliased)')
plt.title('Sampling below Nyquist Rate (Aliasing)')
plt.legend()

plt.tight_layout()
plt.show()
````

##  Explanation: Sampling and Aliasing

### **1. Sampling**

Sampling is the process of converting a continuous-time signal (like a sine wave) into a discrete-time signal by measuring its value at regular intervals (sampling rate).

* If the sampling rate is **high enough** (above twice the highest frequency component of the signal), we can **perfectly reconstruct** the original signal.
* This is known as the **Nyquist theorem**.

### **2. Aliasing**

Aliasing occurs when a signal is sampled **too slowly** (below the Nyquist rate). In this case:

* High-frequency components appear as **lower frequencies** in the sampled signal.
* The result is a **distorted version** of the original signal.

````python
import numpy as np 
import matplotlib.pyplot as plt

# Original signal frequency
f = 10  # Hz

# Sampling frequencies
f1 = 15  # Hz (Below Nyquist  aliasing expected)
f2 = 30  # Hz (Above Nyquist  no aliasing)

# Continuous time vector for 1 second, high resolution
t = np.linspace(0, 1, 1000, endpoint=False)
x = np.sin(2 * np.pi * f * t)  # Original continuous signal

# Sampling time points
t1 = np.arange(0, 1, 1 / f1)  # Low sampling rate (15 Hz)
t2 = np.arange(0, 1, 1 / f2)  # High sampling rate (30 Hz)

# Sampled signals (at original signal frequency f, not sampling frequency)
x1 = np.sin(2 * np.pi * f * t1)
x2 = np.sin(2 * np.pi * f * t2)

# Plotting
plt.figure(figsize=(12, 8))

# Plot 1: Original continuous signal
plt.subplot(3, 1, 1)
plt.plot(t, x, label='Original Signal (10 Hz)', color='blue')
plt.title("Original Continuous-Time Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.axhline(0, color="red", ls='--')  # horizontal axis
plt.axvline(0, color="red", ls='--')  # vertical axis
plt.legend()
plt.grid(True)

# Plot 2: Sampled below Nyquist rate (aliasing)
plt.subplot(3, 1, 2)
plt.stem(t1, x1, linefmt='C1-', markerfmt='C1o', basefmt=" ",  label="15 Hz Sampled Signal")
plt.plot(t, x, 'k--', alpha=0.3, label='Original (for reference)')
plt.title("Sampled at 15 Hz (Below Nyquist  Aliasing)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)

# Plot 3: Sampled above Nyquist rate (no aliasing)
plt.subplot(3, 1, 3)
plt.stem(t2, x2, linefmt='C2-', markerfmt='C2o', basefmt=" ",  label="30 Hz Sampled Signal")
plt.plot(t, x, 'k--', alpha=0.3, label='Original (for reference)')
plt.title("Sampled at 30 Hz (Above Nyquist  Accurate)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Original signal parameters
f_signal = 10  # Hz
T = 1          # seconds

# Time vectors
continuous_time = np.linspace(0, T, 1000)

# High sampling rate (no aliasing)
high_fs = 50  # Hz
high_sampled_time = np.linspace(0, T, int(high_fs*T))

# Low sampling rate (aliasing occurs)
low_fs = 15  # Hz
low_sampled_time = np.linspace(0, T, int(low_fs*T))

# Generate signals
continuous_signal = np.sin(2 * np.pi * f_signal * continuous_time)

high_sampled_signal = np.sin(2 * np.pi * f_signal * high_sampled_time)

low_sampled_signal = np.sin(2 * np.pi * f_signal * low_sampled_time)

# Aliased signal - Compute using low-frequency sampled time
f_alias = abs(low_fs - f_signal)  # 5 Hz (alias frequency)
aliased_signal = np.sin(2 * np.pi * f_alias * low_sampled_time)  # Aliased signal at low_sampled_time

# Plotting
plt.figure(figsize=(14, 9))

# Continuous signal
plt.subplot(2, 1, 1)
plt.plot(continuous_time, continuous_signal, label='Original Signal (Continuous)', color='gray')
plt.stem(high_sampled_time, high_sampled_signal, linefmt='C1-', markerfmt='C1o', basefmt=" ", label='High Rate Samples (No Aliasing)')
plt.plot(high_sampled_time, high_sampled_signal)
plt.title('High Sampling Rate (No Aliasing)')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# Aliased signal
plt.subplot(2, 1, 2)
plt.plot(continuous_time, continuous_signal, label='Original Signal (Continuous)', color='gray')
plt.stem(low_sampled_time, low_sampled_signal, label='Low Rate Samples (Aliasing)')
plt.plot(low_sampled_time, aliased_signal, color='orange', label='Aliased Signal')
plt.title('Low Sampling Rate (Aliasing Effect)')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.show()
````
