```python
import numpy as np
import matplotlib.pyplot as plt

# Original signal: x_a(t) = 3cos(100πt)
A = 3
f_analog = 50  # Hz (from 100π = 2πf → f = 50 Hz)

# Sampling rates
fs1 = 200  # Hz (No aliasing)
fs2 = 75   # Hz (Aliasing)

# Time duration to visualize
t_max = 0.1  # 100 ms

# Sample indices
n1 = np.arange(0, t_max, 1/fs1)
n2 = np.arange(0, t_max, 1/fs2)

# Discrete-time signals (realization after sampling)
x1 = A * np.cos(2 * np.pi * f_analog * n1)  # for fs = 200 Hz
x2 = A * np.cos(2 * np.pi * f_analog * n2)  # for fs = 75 Hz

# Plotting the sampled signals
plt.figure(figsize=(10, 5))

# fs = 200 Hz
plt.subplot(2, 1, 1)
plt.stem(n1, x1)
plt.title("Sampled at 200 Hz (No Aliasing)")
plt.ylabel("x[n]")
plt.grid(True)

# fs = 75 Hz
plt.subplot(2, 1, 2)
plt.stem(n2, x2)
plt.title("Sampled at 75 Hz (Aliasing Occurs)")
plt.xlabel("Time (s)")
plt.ylabel("x[n]")
plt.grid(True)

plt.tight_layout()
plt.show()
```
