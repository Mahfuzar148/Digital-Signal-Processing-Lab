````python
import numpy as np
import matplotlib.pyplot as plt

# ---------- Part 1: Single Frequency Signal ----------
def sample_single_tone(f_s):
    t = np.arange(0, 0.04, 1e-4)  # Continuous time
    x_t = 3 * np.cos(100 * np.pi * t)

    n = np.arange(0, 0.04, 1/f_s)
    x_n = 3 * np.cos(100 * np.pi * n)

    plt.figure(figsize=(10, 3))
    plt.plot(t, x_t, label='Original Signal (Analog)', alpha=0.5)
    plt.stem(n, x_n, linefmt='r-', markerfmt='ro', basefmt='k-', label=f'Sampled at {f_s} Hz')
    plt.title(f'x(t) = 3cos(100πt) Sampled at {f_s} Hz')
    plt.xlabel('Time (s)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# ---------- Part 2: Composite Signal Sampling ----------
def sample_composite_signal(f_s):
    t = np.arange(0, 0.01, 1e-5)  # Higher res for analog
    x_t = 3*np.cos(200*np.pi*t) + 5*np.sin(600*np.pi*t) + 10*np.cos(1200*np.pi*t)

    n = np.arange(0, 0.01, 1/f_s)
    x_n = 3*np.cos(200*np.pi*n) + 5*np.sin(600*np.pi*n) + 10*np.cos(1200*np.pi*n)

    plt.figure(figsize=(10, 3))
    plt.plot(t, x_t, label='Original Signal', alpha=0.5)
    plt.stem(n, x_n, linefmt='g-', markerfmt='go', basefmt='k-', label=f'Sampled at {f_s} Hz')
    plt.plot(n,x_n, label='Sampled Signal', color='orange', linestyle='--', alpha=0.7)
    plt.title(f'Composite Signal Sampled at {f_s} Hz')
    plt.xlabel('Time (s)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Run Part 1 for 200 Hz and 75 Hz
sample_single_tone(200)
sample_single_tone(75)

# Run Part 2 for multiple sampling rates
for fs in [1200, 1000, 800,600]:
    sample_composite_signal(fs)
````
