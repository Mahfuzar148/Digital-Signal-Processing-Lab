```python
import numpy as np
import matplotlib.pyplot as plt

# Task 1: Signal Smoothing with First-Order Difference Equation
x = np.array([1, 2, 2, 10, 2, 2, 1])  # Input signal
alpha_values = [0.2, 0.8]  # Alpha values

# Function to apply the difference equation
def smooth_signal(x, alpha):
    y = np.zeros(len(x))
    for n in range(len(x)):
        y[n] = alpha * x[n] if n == 0 else (1 - alpha) * y[n - 1] + alpha * x[n]
    return y

# Plot for Task 1
plt.figure(figsize=(10, 6))
for alpha in alpha_values:
    plt.plot(smooth_signal(x1, alpha), label=f'Alpha = {alpha}')
plt.plot(x, 'k--', label='Original Signal', alpha=0.5)
plt.title('Signal Smoothing using First-Order Difference Equation')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()

# Task 2: Sinusoidal Signal with Random Noise
N = 100  # Number of samples
n = np.arange(N)
x2 = np.sin(2 * np.pi * n / N)  # Clean sine wave
x2_noisy = x2 + np.random.randn(N)  # Noisy signal

# Plot for Task 2
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(x2, label='Original Clean Signal')
plt.title('Original Clean Sine Wave')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(x2_noisy, label='Noisy Signal', color='r')
plt.title('Noisy Signal')
plt.grid(True)

plt.subplot(3, 1, 3)
for alpha in alpha_values:
    plt.plot(smooth_signal(x2_noisy, alpha), label=f'Smoothed (Alpha = {alpha})')
plt.title('Smoothed Signals for Noisy Input')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

```
