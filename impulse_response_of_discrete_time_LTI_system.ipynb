```python
import numpy as np
import matplotlib.pyplot as plt

# Define length for simulation
N = 20  # We'll compute from n=0 to n=19

# Define unit step function u[n]
def unit_step(n):
    return np.where(n >= 0, 1, 0)

# Create discrete time vector
n = np.arange(0, N)

# Define input x[n] = u[n]
x = unit_step(n)

# Define h[n] = u[n] - u[n-5]
h = unit_step(n) - unit_step(n - 5)

# Compute convolution y[n] = x[n] * h[n]
y = np.convolve(x, h)

# Time axis for convolution output
n_conv = np.arange(0, len(y))

# Plotting all signals
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.stem(n, x, basefmt=" ")
plt.title('Input x[n] = u[n]')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.stem(n, h, basefmt=" ")
plt.title('Impulse Response h[n] = u[n] - u[n-5]')
plt.xlabel('n')
plt.ylabel('h[n]')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.stem(n_conv, y, basefmt=" ")
plt.title('Output y[n] = x[n] * h[n]')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.grid(True)

plt.tight_layout()
plt.show()
# This code simulates the impulse response of a discrete-time LTI system by convolving a unit step input with a defined impulse response.
# It generates plots for the input signal, impulse response, and output signal, providing a clear visualization of the system's behavior.
# The use of `np.convolve` allows for efficient computation of the convolution, and the `unit_step` function defines the unit step signal used in the input and impulse response.
# The plots are generated using `matplotlib`, which provides a clear visual representation of the signals involved.
# The output will show the input signal, the impulse response, and the resulting output signal after convolution.
```
