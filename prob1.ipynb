```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for n (time index)
n = np.arange(-10, 11, 1)

# 1. Unit Step Signal (u(n))
# The unit step signal is defined as u(n) = 1 for n >= 0, and u(n) = 0 for n < 0
u_n = np.where(n >= 0, 1, 0)

# 2. Ramp Signal (r(n))
# The ramp signal is defined as r(n) = n for n >= 0, and r(n) = 0 for n < 0
r_n = np.where(n >= 0, n, 0)

# 3. Exponential Signal (x(n) = 0.5^n)
# The exponential signal is defined as x(n) = 0.5^n
x_n = 0.5 ** n

# 4. Sine Signal (x(n) = sin(2*pi*0.1*n))
# The sine signal is a periodic signal with frequency 0.1 Hz
sin_n = np.sin(2 * np.pi * 0.1 * n)

# 5. Cosine Signal (x(n) = cos(2*pi*0.1*n))
# The cosine signal is also periodic with frequency 0.1 Hz
cos_n = np.cos(2 * np.pi * 0.1 * n)

# Plotting all the signals
plt.figure(figsize=(10, 8))

# Plot Unit Step Signal
plt.subplot(3, 2, 1)
plt.stem(n, u_n, use_line_collection=True)
plt.title('Unit Step Signal (u(n))')
plt.xlabel('n')
plt.ylabel('u(n)')
plt.grid(True)

# Plot Ramp Signal
plt.subplot(3, 2, 2)
plt.stem(n, r_n, use_line_collection=True)
plt.title('Ramp Signal (r(n))')
plt.xlabel('n')
plt.ylabel('r(n)')
plt.grid(True)

# Plot Exponential Signal
plt.subplot(3, 2, 3)
plt.stem(n, x_n, use_line_collection=True)
plt.title('Exponential Signal (x(n) = 0.5^n)')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.grid(True)

# Plot Sine Signal
plt.subplot(3, 2, 4)
plt.plot(n, sin_n)
plt.title('Sine Signal (x(n) = sin(2*pi*0.1*n))')
plt.xlabel('n')
plt.ylabel('sin(n)')
plt.grid(True)

# Plot Cosine Signal
plt.subplot(3, 2, 5)
plt.plot(n, cos_n)
plt.title('Cosine Signal (x(n) = cos(2*pi*0.1*n))')
plt.xlabel('n')
plt.ylabel('cos(n)')
plt.grid(True)

# Adjust layout to prevent overlap of subplots
plt.tight_layout()

# Show the plot
plt.show()
```
