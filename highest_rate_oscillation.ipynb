```python
import numpy as np 
import matplotlib.pyplot as plt 

omega1 = np.pi / 4  # Frequency of the first oscillation
omega2 = 3 * np.pi / 4  # Frequency of the second oscillation
omega3 = np.pi/2  # Frequency of the third oscillation
omega4 = np.pi  # Frequency of the fourth oscillation



n = np.arange(0, 100)  # Sample indices
x1 = np.cos(omega1 * n)  # First oscillation

x2 = np.cos(omega2 * n)  # Second oscillation
x3 = np.cos(omega3 * n)  # Third oscillation
x4 = np.cos(omega4 * n)  # Fourth oscillation

plt.figure(figsize=(18,10))

plt.subplot(4, 1, 1)
plt.plot(n,x1, 'C0-')
plt.stem(n, x1, linefmt='C0-', markerfmt='C0o', basefmt=" ", label='1st Oscillation (/4)')
plt.legend()


plt.subplot(4, 1, 2)
plt.plot(n,x2, 'C1-')
plt.stem(n, x2, linefmt='C0-', markerfmt='C0o', basefmt=" ", label='1st Oscillation (3/4)')
plt.legend()


plt.subplot(4, 1, 3)
plt.plot(n,x3, 'C0-')
plt.stem(n, x3, linefmt='C0-', markerfmt='C0o', basefmt=" ", label='1st Oscillation (/4)')
plt.legend()

plt.subplot(4, 1, 4)
plt.plot(n,x4, 'C0-')
plt.stem(n, x4, linefmt='C0-', markerfmt='C0o', basefmt=" ", label='1st Oscillation (/4)')
plt.legend()



plt.tight_layout()
plt.show()

```
